Feature Selection - Droping constant features

```python
import pandas as pd
```

```python
data = pd.DataFrame({'A':[1,2,3,4,5,6],
                     'B':[7,8,9,10,11,12],
                     'C':[0,0,0,0,0,0],
                     'D':[1,1,1,1,1,1]})
data
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
selectionfunction = VarianceThreshold(threshold=0)
selectionfunction.fit(data)
```

```python
selectionfunction.get_support()
```

```python
data.columns[selectionfunction.get_support()]
```

```python
constant_columns = [column for column in data.columns
                    if column not in data.columns[selectionfunction.get_support()]]

constant_columns
```

```python
data.drop(constant_columns,axis=1)
```

```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
```

```python
x = df.drop(labels='TARGET', axis = 1)
y = df['TARGET']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=0)
X_train.shape, X_test.shape
```

Lets apply Variance Threshold algo

```python
var_thres = VarianceThreshold(threshold=0)
var_thres.fit(X_train)
```

```python
sum(var_thres.get_support())
```

```python
var_thres.get_support()
```

```python
len(X_train.columns[var_thres.get_support()])
```

```python
constant_columns = [column for column in X_train.columns
                    if column not in X_train.columns[var_thres.get_support()]]

print(len(constant_columns))
```

```python
for columns in constant_columns:
  print(columns)
```

```python
X_train.drop(constant_columns, axis=1)
```
